# Task 1 — Data Parsing & Profile Cleaning

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
cleaned_students = []                       #----------------------> new list to store clean data

for student in raw_students:

    name = student["name"].strip().title()   # ------------------------> Clean the  names by stripping and title case

    roll = int(student["roll"])             # --------------------------> Convert roll to integer

    marks = []                              # ------->create new list for storining integer marks
    split_marks = student["marks_str"].split(",")

    for m in split_marks:                 #-----------> loop into the uncleaned marks data to convert it into interger and store into a new list.
        cleaned_mark = m.strip()
        number = int(cleaned_mark)
        marks.append(number)

    words = name.split()                    # --------------------------> to validate each name name
    is_valid = True

    for word in words:                      #-------------------------> loop into each word of the name to check if it contains only alphabetic characters
        if not word.isalpha():
            is_valid = False
            break

    if is_valid:                            #--------------------------> Print validation result
        print(f"{name} ✓ Valid name")     
    else:
        print(f"{name} ✗ Invalid name")

    cleaned_students.append({                  #-----------------------------> Store cleaned data
        "name": name,
        "roll": roll,
        "marks": marks
    })

    print("=" * 40)                       # ----------------------------------> Print formatted profile
    print(f"Student : {name}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("=" * 40)



for student in cleaned_students:          # -----------------------------------> Find student with roll no.  103
    if student["roll"] == 103:
        print("UPPERCASE:", student["name"].upper())
        print("lowercase:", student["name"].lower())
```

# Task 2 — Marks Analysis Using Loops & Conditionals

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
print(f"\nStudent: {student_name}")
print("=" * 40)


#  1. Print subject + marks + grade


for i in range(len(subjects)):          #  ------------------------------> Loop through subjects and marks using index
    subject = subjects[i]
    mark = marks[i]

    # Grade logic
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subject} : {mark} → Grade {grade}")


    

# 2. Calculations



total = 0                           #------------------------------> Initialize total for average calculation
highest = marks[0]              #------------------------------> Initialize highest and lowest with the first mark for comparison
lowest = marks[0]
high_subject = subjects[0]          #------------------------------> Initialize corresponding subjects for highest and lowest marks
low_subject = subjects[0]

for i in range(len(marks)):         #------------------------------> Loop through marks using index for calculations
    total += marks[i]

    if marks[i] > highest:          #------------------------------> Check for highest mark and update if current mark is higher
        highest = marks[i]
        high_subject = subjects[i]

    if marks[i] < lowest:           #------------------------------> Check for lowest mark and update if current mark is lower
        lowest = marks[i]
        low_subject = subjects[i]

average = total / len(marks)        #------------------------------> Calculate average by dividing total by number of marks

print("\n" + "=" * 40)
print(f"Total Marks   : {total}")
print(f"Average Marks : {round(average, 2)}")
print(f"Highest       : {high_subject} ({highest})")
print(f"Lowest        : {low_subject} ({lowest})")




# 3. While loop for new entries
new_count = 0

while True:
    subject = input("\nEnter subject name (or 'done' to stop): ")

    if subject.lower() == "done":               #------------------------------> .lower() to make the input case-insensitive
        break                                   #------------------------------> Exit loop if user is done

    mark_input = input("Enter marks (0–100): ")     #------------------------------> Take marks input as string for validation

    # Validate input
    if not mark_input.isdigit():                    #------------------------------> Check if input is a number
        print("Invalid input! Must be a number.")
        continue

    mark = int(mark_input)                  #------------------------------> Convert valid input to integer for range validation

    if mark < 0 or mark > 100:              #------------------------------> Check if marks are within valid range
        print("Invalid marks! Must be between 0 and 100.")
        continue

    # Add valid data
    subjects.append(subject)                #------------------------------> Add new subject to the list
    marks.append(mark)                      #  ------------------------------> Add new mark to the list
    new_count += 1

print("\n" + "=" * 40)
print(f"New subjects added: {new_count}")       #------------------------------> Print count of new entries

# Updated average after adding new entries
total = 0
for m in marks:
    total += m

updated_avg = total / len(marks)

print(f"Updated Average: {round(updated_avg, 2)}")
```

# Task 3 — Class Performance Summary 

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
print("Name              | Average | Status")
print("--------------------------------------")

#1. Initialize pass and fail counts, top student tracking, and total average sum for class average calculation

pass_count = 0                      
fail_count = 0
top_name = ""
top_avg = 0
total_avg_sum = 0

#2. Loop through each student and calculate average, pass/fail status, track topper, and total average.

for student in class_data:
    name = student[0]
    marks = student[1]

    total = 0
    for m in marks:
        total += m

    avg = round(total / len(marks), 2)

    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    if avg > top_avg:
        top_avg = avg
        top_name = name

    total_avg_sum += avg

    print(f"{name:<18} | {avg:<7} | {status}")

class_avg = total_avg_sum / len(class_data) 

print("\nSummary")
print("--------")
print("Passed:", pass_count)
print("Failed:", fail_count)
print(f"Topper: {top_name} ({top_avg})")
print(f"Class Average: {round(class_avg, 2)}")
```

```python
essay = "   python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.   "

# Step 1: Strip spaces
clean_essay = essay.strip()
print("Step 1:", clean_essay)

# Step 2: Convert to Title Case
title_essay = clean_essay.title()
print("Step 2:", title_essay)

# Step 3: Count "python"
count = clean_essay.count("python")
print("Step 3: 'python' appears", count, "times")

# Step 4: Replace "python" with "Python 🐍"
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("Step 4:", replaced_essay)

#  Step 5: Split into sentences
sentences = clean_essay.split(". ")
print("Step 5:", sentences)

#  Step 6: Print numbered sentences
print("Step 6:")
for i in range(len(sentences)):             #-------------> Loop through sentences using index to print numbered list
    sentence = sentences[i]

    # Add "." if missing
    if not sentence.endswith("."):
        sentence = sentence + "."

    print(f"{i+1}. {sentence}")
```
